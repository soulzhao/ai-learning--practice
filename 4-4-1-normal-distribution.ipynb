正态分布函数学习

numpy.random.normal
参数1: loc 正态分布的中心，均值（Mean (“centre”) of the distribution）

参数2: scale  正态分布的标准差 Standard deviation (spread or “width”) of the distribution. Must be non-negative.

参数3: size Output shape[int or tuple of ints, optional]


```python
import numpy as np
mu = 0.0
sigma = 0.1
s = np.random.normal(loc=mu, scale=sigma, size=1000)
```

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```
