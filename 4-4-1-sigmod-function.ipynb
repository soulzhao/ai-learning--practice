|  函数名称   | 英文名称  |  中文名称  |
|  ----  | ----  |  ----  |
| logit(x[, out])  | Logit ufunc for ndarrays. | 用于将概率转换为对数几率。它的输入是一个概率值（0到1之间），输出是对数几率值（负无穷到正无穷之间）。这个函数在逻辑回归模型中经常用于计算预测结果的对数几率。|
| expit(x[, out])  | Expit (a.k.a.) | 逻辑斯蒂函数的标准形式，也被称为sigmoid函数。它将输入值映射到一个介于0和1之间的概率值。逻辑斯蒂函数在机器学习中被广泛应用，特别是在二分类问题中。它可以将线性模型的输出转化为概率形式。expit is the inverse of logit|
| log_expit(x[, out])  | Logarithm of the logistic sigmoid function. | 它首先计算逻辑斯蒂函数的值，然后再取对数。这个函数在某些情况下可以用于数值稳定性，避免浮点数溢出或下溢。|

```python
import numpy as np
from scipy.special import logit, expit, log_expit

import matplotlib.pyplot as plt
x = np.linspace(0, 1, 501)
y = logit(x)
plt.plot(x, y)
plt.grid()
plt.ylim(-6, 6)
plt.xlabel('x')
plt.title('logit(x)')
plt.show()
```

```python
x = np.linspace(-6, 6, 121)
y = expit(x)
plt.plot(x, y)
plt.grid()
plt.xlim(-6, 6)
plt.xlabel('x')
plt.title('expit(x)')
plt.show()
```
